```python
# read. mat-file
each_mat = 'Testset_Result_Vit_epoch0500.mat'

# read out the struct of .mat
array_struct = scio.loadmat(each_mat)
```

```python
# read out the arry we neeed from the struct
array_TrainLoss = np.squeeze(array_struct['TrainLoss'])
array_ValidLoss = np.squeeze(array_struct['ValidLoss'])
array_TestLoss = np.squeeze(array_struct['TestLoss'])
```

```python
array_data0 = array_struct['GT'] 
array_data1 = array_struct['result'] 
```

```python
# Mean Squared ErrorMSE
# MSE =  (predicted_value - true_value)^2 / number
MSE_list = []
for i in range(0, array_data0.shape[0], 1): 
    Sum_MSE = 0
    for j in range(0, array_data0.shape[1], 1):
        true_value = array_data0[i, j]
        predicted_value = array_data1[i, j]
        SQ = (true_value - predicted_value) ** 2
        Sum_MSE += SQ
    MSE = Sum_MSE / array_data0.shape[1]
    MSE_list.append(MSE)

plt.plot(MSE_list)            
plt.ylabel('MSE')

MSE_array = np.array(MSE_list)
print(MSE_array.shape)

a = np.mean(MSE_array)
s = np.std(MSE_array)

print(str(a) + u"\u00B1" + str(s))
```

```python
# Root Mean Squared ErrorRMSE
# RMSE = MSE
RMSE_list = []
for i in range(0, array_data0.shape[0], 1): 
    Sum_MSE = 0
    for j in range(0, array_data0.shape[1], 1):
        true_value = array_data0[i, j]
        predicted_value = array_data1[i, j]
        SQ = (true_value - predicted_value) ** 2
        Sum_MSE += SQ
    RMSE = math.sqrt(Sum_MSE / array_data1.shape[1])
    RMSE_list.append(RMSE)

plt.plot(RMSE_list) 
plt.ylabel('RMSE')

RMSE_array = np.array(RMSE_list)


a = np.mean(RMSE_array)
s = np.std(RMSE_array)

print(str(a) + u"\u00B1" + str(s))
```

```python
# Mean Absolute ErrorMAE
# MAE =  |predicted_value - true_value | / number
MAE_list = []
for i in range(0, array_data0.shape[0], 1): 
    Sum_MSE = 0
    for j in range(0, array_data1.shape[1], 1):
        true_value = array_data0[i, j]
        predicted_value = array_data1[i, j]
        SQ = abs(true_value - predicted_value)
        Sum_MSE += SQ
    MAE = Sum_MSE / array_data1.shape[1]
    MAE_list.append(MAE)

    
plt.plot(MAE_list) 
plt.ylabel('MAE')

MAE_array = np.array(MAE_list)
print(MAE_array.shape)

a = np.mean(MAE_array)
s = np.std(MAE_array)

print(str(a) + u"\u00B1" + str(s))
```
